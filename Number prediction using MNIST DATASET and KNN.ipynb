```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
digits = load_digits()
x = digits.data
y = digits.target
print('dimension de x:', x.shape)
```

```python
print('dimension de y:', y.shape)
```

```python
y
```

```python
plt.imshow(digits['images'][0],cmap='Greys_r')
```

```python
x_tarin,x_test,y_tarin,y_test=train_test_split(x,y,test_size=0.2)
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_tarin,y_tarin)
```

```python
score=knn.score(x_test,y_test)
score
```

```python
neighbors = np.arange(1,21)
train_scores=np.empty(len(neighbors))
test_scores=np.empty(len(neighbors))
for i,k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_tarin,y_tarin)
    train_scores[i]=knn.score(x_tarin,y_tarin)
    test_scores[i]=knn.score(x_test,y_test);
print(test_scores)
    
```

```python
plt.plot(test_scores, 'o-')
plt.title('K-NN precision par nombre de voisins pour le test')
plt.xlabel('Nombre de voisoins')
plt.ylabel('Precision')
plt.show()
```
